```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras import layers
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import Normalizer
import matplotlib.pyplot as plt
from tensorflow.keras.layers import Dense, Dropout, Flatten, Conv2D, MaxPool2D
```

```python
df = pd.read_csv('Preprocessed_dataset.csv')
df.head(10)
```

```python
df.fire_size_class = [ ord(x) - 66 for x in df.fire_size_class ]
df.head(10)
```

```python
df.fire_size_class.unique()
```

```python
#Define the columns to import from the .CSV file
col_list = ["fire_size", "fire_size_class", "stat_cause_descr", "latitude", "longitude", "state", "disc_date_pre", "disc_pre_year", "disc_pre_month", "weather_file", "Temp_pre_30", "Temp_pre_15", "Temp_pre_7", "Wind_pre_30", "Wind_pre_15", "Wind_pre_7", "Hum_pre_30", "Hum_pre_15", "Hum_pre_7", "Prec_pre_30", "Prec_pre_15", "Prec_pre_7", "remoteness"]

#Define the targets to use for the models
target_list = ["fire_size_class"]
#feature_list = ["stat_cause_descr", "latitude", "longitude", "disc_date_pre", "disc_pre_year", "disc_pre_month", "Temp_pre_30", "Temp_pre_15", "Temp_pre_7", "Wind_pre_30", "Wind_pre_15", "Wind_pre_7", "Hum_pre_30", "Hum_pre_15", "Hum_pre_7", "Prec_pre_30", "Prec_pre_15", "Prec_pre_7", "remoteness"]
feature_list = ["latitude", "longitude", "Temp_pre_30", "Temp_pre_15", "Temp_pre_7", "Wind_pre_30", "Wind_pre_15", "Wind_pre_7", "Hum_pre_30", "Hum_pre_15", "Hum_pre_7", "Prec_pre_30", "Prec_pre_15", "Prec_pre_7", "remoteness", "Arson","Campfire", "Children", "Debris Burning", "Equipment Use", "Fireworks", "Lightning", "Miscellaneous", "Missing/Undefined", "Powerline", "Railroad", "Smoking", "Structure"]
```

```python
df.drop('weather_file', axis=1)
targets = df[target_list]
features = df[feature_list]
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.5)
```

```python
transformer = Normalizer().fit(X_train)
X_train = transformer.transform(X_train)
X_test = transformer.transform(X_test)
```

```python
X_train = tf.cast(X_train, tf.float32)
y_train = tf.cast(y_train, tf.float32)
X_test = tf.cast(X_test, tf.float32)
y_test = tf.cast(y_test, tf.float32)
```

```python
layers = [Dense(100, activation="relu"),
          Dropout(0.2),
          Dense(100, activation="relu"),
          Dropout(0.2),
          Dense(6, activation="softmax")]


model = tf.keras.Sequential(layers)


model.compile(optimizer="adam", loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=5)
```
